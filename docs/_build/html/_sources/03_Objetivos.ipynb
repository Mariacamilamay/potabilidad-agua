## 3 Objetivos

3.1 Objetivo General

Desarrollar un modelo de aprendizaje automático que permita predecir la potabilidad del agua a partir de sus características fisicoquímicas, mediante el análisis y procesamiento de datos.

3.2 Objetivos Especificos

1 Realizar un análisis exploratorio de los datos para comprender la distribución y el comportamiento de las variables relacionadas con la potabilidad del agua.

2 Identificar las variables con mayor influencia en la clasificación del agua como potable o no potable.

3 Preparar y transformar los datos para su utilización en modelos de aprendizaje automático.

4 Entrenar y evaluar modelos de clasificación para predecir la potabilidad del agua.

5 Comparar el desempeño de los modelos utilizados y seleccionar el que presente los mejores resultados.
