# 1. Introducción

La calidad del agua es un factor esencial para la salud y el bienestar de las personas. Evaluar si el agua es apta para el consumo humano requiere analizar diversas características fisicoquímicas que influyen en su potabilidad.

En este proyecto se realiza un análisis exploratorio de datos sobre muestras de agua con el fin de identificar patrones y relaciones entre las variables. Posteriormente, se aplicarán técnicas de aprendizaje automático para desarrollar un modelo capaz de predecir si una muestra de agua es potable o no a partir de sus características.
