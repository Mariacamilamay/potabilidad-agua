## 4 Justificación

El acceso a agua potable es fundamental para la salud y la calidad de vida de la población. Por ello, contar con herramientas que permitan evaluar de manera eficiente la calidad del agua es de gran importancia para la toma de decisiones y la prevención de riesgos asociados al consumo de agua no apta.

Este proyecto aplica técnicas de análisis de datos y aprendizaje automático para identificar los factores que influyen en la potabilidad del agua y desarrollar un modelo predictivo capaz de clasificar nuevas muestras. De esta manera, se busca demostrar cómo la ciencia de datos puede contribuir a la evaluación de la calidad del agua mediante el uso de información basada en datos.
