## 2 Planteamiento del Problema

La calidad del agua es un aspecto fundamental para la salud humana, ya que el consumo de agua contaminada puede generar diversas enfermedades y afectar la calidad de vida de la población. Determinar si una fuente de agua es potable requiere analizar diferentes características fisicoquímicas, proceso que puede resultar complejo debido a la cantidad de variables involucradas y a las relaciones existentes entre ellas.

Ante esta situación, surge la necesidad de identificar cuáles son los factores que influyen en la potabilidad del agua y evaluar si es posible predecir esta condición mediante técnicas de análisis de datos y aprendizaje automático. Por ello, este proyecto busca responder la siguiente pregunta:

¿Es posible predecir la potabilidad del agua a partir de sus características fisicoquímicas mediante el uso de modelos de aprendizaje automático?
